```python
import pandas as pd
pd.set_option('display.max_columns', None)# pandas show all columns of table instead of restricted
#pd.set_option('display.max_rows', None)
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as px

from envirocar import TrackAPI, DownloadClient, BboxSelector, ECConfig
from envirocar import inspection as inspect
from envirocar import correction as correct
from envirocar import manipulation as manipulate

# create an initial but optional config and an api client
config = ECConfig()
track_api = TrackAPI(api_client=DownloadClient(config=config))

```

```python
bbox = BboxSelector([
    7.554130554199218, # min_x
    51.95590322041212, # min_y
    7.590351104736328, # max_x
    51.97874790276371  # max_y
])
# issue a query
df_tracks = track_api.get_tracks(bbox=bbox, num_results=20) 
```

First clean the data a little bit as by dropping dublicated rows and dropping all unit columns

```python
correct.drop_dublicates(df_tracks)
```

As we are not interested in the unit columns (which only hold the unit of measurements) we drop them from the dataframe

```python
manipulate.drop_unit_columns(df_tracks).head()
```

Now inspect the datatypes of the variables

```python
inspect.plot_tracks(inspect.get_single_track(df_tracks, '5f0ef89c00375c5a2641ef84'),'Engine Load.value' )
```

```python
df_tracks.dtypes
```

Here you need to make a decision which data for your analysis is crucial.  
In general you may impute data which have about 25-30% missing values, however, if you can guess the distribution
of the specific column or the variable is especially important, or you have variables which is appropriate for prediction (and other cases) you may also impute data which have more missing values.  
We will do here a simple univariate interpolation with k-nearest.  

### Inspect proportion missing values
With the following command all columns are shown which have less than 60% missing values, and additionally, we drop all variables from the dataframe which provide less than 60% percent data.

```python
inspect.missing_values_per_variable(df_tracks, 60, dropCol=True)
```

### Missing values per track
We should also have a look at missing values per track.

```python
count, proportion = inspect.missing_values_per_track(df_tracks)
proportion
```
